```python
# ==============================================================================
# SECURE PLAIN-TEXT PROTOCOL (ZERO-JSON BACKEND MATRIX)
# ==============================================================================
from IPython.display import HTML, display
from google.colab import output

class PlainSessionState:
    def __init__(self):
        self.intention = "Unassigned"
        self.confirmed_ideas = []
        self.confirmed_blockers = []

if 'plain_state' not in globals():
    plain_state = PlainSessionState()

def py_chat_pure_string(user_message):
    """Bypasses JSON parsing entirely using plain text pipe extraction"""
    user_message = user_message.strip()
    if not user_message:
        return "STATUS:EMPTY"

    processed_text = user_message.upper()
    ai_response = f"[Ping Test] Converted input to uppercase: {processed_text}"

    insight_type = "none"
    insight_content = ""

    if "IDEA" in processed_text:
        insight_type = "idea"
        insight_content = f"Extracted Innovation: {user_message}"
    elif "RISK" in processed_text or "BLOCKER" in processed_text:
        insight_type = "blocker"
        insight_content = f"Extracted Risk Constraint: {user_message}"

    # Structured delimiter string output instead of volatile JSON token syntax
    return f"{ai_response}|||{insight_type}|||{insight_content}"

def py_store_pure_string(payload):
    """Processes pipeline updates via custom dynamic token splitting"""
    try:
        parts = payload.split("|||")
        data_type = parts[0]
        text_content = parts[1]

        if data_type == "idea":
            plain_state.confirmed_ideas.append(text_content)
        elif data_type == "blocker":
            plain_state.confirmed_blockers.append(text_content)
        elif data_type == "intent":
            plain_state.intention = text_content
        return "STATUS:SUCCESS"
    except Exception as e:
        return f"STATUS:ERROR:{str(e)}"

def py_compile_pure_string():
    context_summary = (
        f"Partner Strategy Track: {plain_state.intention}\n"
        f"Stored Ideas: {'; '.join(plain_state.confirmed_ideas)}\n"
        f"Logged Risks: {'; '.join(plain_state.confirmed_blockers)}"
    )
    return f"# STRATEGIC BRIEF\n\n{context_summary}"

# EXPLICIT CALLBACK REGISTER MAPPING
output.register_callback('py_chat_pure_string', py_chat_pure_string)
output.register_callback('py_store_pure_string', py_store_pure_string)
output.register_callback('py_compile_pure_string', py_compile_pure_string)

# FRONTEND STACK WITH RAW TEXT PIPELINES
html_blueprint = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <script src="https://cdn.tailwindcss.com"></script>
</head>
<body class="bg-slate-950 text-slate-100 font-sans flex flex-col h-[600px] w-full border border-slate-800 rounded-xl overflow-hidden">
    <header class="bg-slate-900 border-b border-slate-800 px-6 py-3 shrink-0">
        <span class="text-base font-bold text-emerald-400">⚡ Plain-Text Secure Communication Bridge</span>
    </header>
    <main class="flex-1 flex overflow-hidden">
        <section class="w-1/2 border-r border-slate-800 flex flex-col p-4 space-y-4">
            <div class="bg-slate-900 p-3 rounded-lg border border-slate-800 text-xs">
                <span class="text-slate-500 block font-bold mb-1">Active Track</span>
                <span id="active-intent" class="text-slate-300 italic">None</span>
            </div>
            <div id="chat-box" class="flex-1 bg-slate-900/40 border border-slate-800 rounded-lg p-3 overflow-y-auto text-xs space-y-2 font-mono"></div>
            <div class="flex space-x-2">
                <input type="text" id="user-input" placeholder="Include word 'idea' or 'risk' to test cards" class="flex-1 bg-slate-950 border border-slate-700 rounded-lg px-3 py-2 text-xs text-slate-100 focus:outline-none" />
                <button id="send-btn" class="bg-emerald-600 px-4 py-2 rounded-lg text-xs font-bold">Send</button>
            </div>
        </section>
        <section class="w-1/2 p-4 flex flex-col space-y-4 bg-slate-900/10">
            <div class="bg-slate-900 p-3 rounded-lg border border-slate-800 text-xs flex-1">
                <span class="text-emerald-400 font-bold block mb-2">[Approved Ideas]</span>
                <div id="ideas-output" class="space-y-1"></div>
            </div>
            <div class="bg-slate-900 p-3 rounded-lg border border-slate-800 text-xs flex-1">
                <span class="text-amber-400 font-bold block mb-2">[Logged Risks]</span>
                <div id="risks-output" class="space-y-1"></div>
            </div>
            <button id="compile-btn" class="bg-indigo-600 py-2 rounded-lg text-xs font-bold shrink-0">Compile Brief</button>
            <div id="brief-output" class="bg-slate-950 p-2 rounded border border-slate-800 font-mono text-[10px] h-24 overflow-y-auto whitespace-pre-wrap"></div>
        </section>
    </main>
    <script>
        (function() {
            let step = 0;
            const chatBox = document.getElementById('chat-box');
            const userInput = document.getElementById('user-input');
            const sendBtn = document.getElementById('send-btn');
            const activeIntent = document.getElementById('active-intent');
            const ideasOutput = document.getElementById('ideas-output');
            const risksOutput = document.getElementById('risks-output');
            const compileBtn = document.getElementById('compile-btn');
            const briefOutput = document.getElementById('brief-output');

            sendBtn.onclick = function() {
                let val = userInput.value.trim();
                if(!val) return;

                logText('USER', val);
                userInput.value = '';

                if (step === 0) {
                    google.colab.kernel.invokeFunction('py_store_pure_string', ['intent|||' + val], {});
                    activeIntent.innerText = val;
                    step = 1;
                }

                google.colab.kernel.invokeFunction('py_chat_pure_string', [val], {})
                    .then(res => {
                        let rawResponse = res.data['text/plain'];
                        let parts = rawResponse.split('|||');
                        let aiReply = parts[0];
                        let insightType = parts[1];
                        let insightContent = parts[2];

                        logText('AI', aiReply);
                        if(insightType !== 'none') {
                            renderCard(insightType, insightContent);
                        }
                    });
            };

            compileBtn.onclick = function() {
                google.colab.kernel.invokeFunction('py_compile_pure_string', [], {})
                    .then(res => {
                        briefOutput.innerText = res.data['text/plain'];
                    });
            };

            function logText(sender, txt) {
                let p = document.createElement('p');
                p.innerText = sender + ": " + txt;
                chatBox.appendChild(p);
                chatBox.scrollTop = chatBox.scrollHeight;
            }

            function renderCard(type, content) {
                let div = document.createElement('div');
                div.className = "p-2 bg-slate-800 border rounded mt-1 space-y-1 border-slate-700";
                div.innerHTML = `<p class='font-bold text-emerald-400'>Authorize \${type}?</p><p class='italic text-slate-300'>"\${content}"</p><button class='bg-emerald-600 px-2 py-0.5 rounded text-[10px] text-white'>Confirm</button>`;
                div.querySelector('button').onclick = function() {
                    google.colab.kernel.invokeFunction('py_store_pure_string', [type + '|||' + content], {});
                    let tgt = type === 'idea' ? ideasOutput : risksOutput;
                    let item = document.createElement('div');
                    item.className = "bg-slate-950 p-1 rounded mt-1 border border-slate-800 text-slate-300";
                    item.innerText = content;
                    tgt.appendChild(item);
                    div.remove();
                };
                chatBox.appendChild(div);
                chatBox.scrollTop = chatBox.scrollHeight;
            }
        })();
    </script>
</body>
</html>
"""
display(HTML(html_blueprint))
```

```python
# ==========================================
# STEP 1: INSTALL REQUIRED LIBRARIES
# ==========================================
!pip install -q transformers accelerate bitsandbytes ipywidgets

import torch
import gc
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import ipywidgets as widgets
from IPython.display import display, clear_output

# ==========================================
# STEP 2: MODEL CONFIGURATION & LOADING (CORRECTED)
# ==========================================
MODEL_ID = "NousResearch/Meta-Llama-3-8B-Instruct"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

print("Loading model and tokenizer... This may take a few minutes.")
# Corrected from parse_pretrained to from_pretrained
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    quantization_config=bnb_config,
    device_map="auto"
)
tokenizer.pad_token = tokenizer.eos_token
print("Model loaded successfully on T4 GPU.")

# ==========================================
# STEP 3: CORE AI INFERENCE FUNCTION
# ==========================================
def process_text_with_llm(user_input):
    """
    Executes Zero-shot Classification and analysis using local Llama-3.
    Enforces memory cleanup to prevent OOM crashes.
    """
    # System prompt enforcing the frozen 3-tag business logic structure
    system_prompt = (
        "You are an elite business analyst. Analyze the user's raw input text and "
        "extract insights strictly into these three frozen categories:\n"
        "1. 💡 Innovation / Solutions (AI application points, automation ideas, process improvements)\n"
        "2. ⚠️ Pain Points / Risks (Operational bottlenecks, human friction, legal/compliance risks)\n"
        "3. 🏗️ Core Entities / Context (Key stakeholders, background, systems involved)\n\n"
        "Format the output using clear Markdown headings and bullet points. Be concise."
    )

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_input}
    ]

    # Apply Llama-3 prompt template
    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    # Execute inference with strict parameter limits
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=1024,
            temperature=0.2, # Low temperature for accurate classification
            top_p=0.9,
            eos_token_id=tokenizer.eos_token_id
        )

    response = tokenizer.decode(outputs[0][inputs.input_ids.shape[1]:], skip_special_tokens=True)

    # CRITICAL: VRAM Safety cleanup to prevent Colab session crash
    del inputs, outputs
    gc.collect()
    torch.cuda.empty_cache()

    return response

# ==========================================
# STEP 4: NATIVE IPYWIDGETS UI INTERFACE (CORRECTED)
# ==========================================
from IPython.display import display, clear_output, Markdown

# UI Components definition
txt_input = widgets.Textarea(
    value='',
    placeholder='Paste your business raw dump here...',
    description='Raw Input:',
    layout=widgets.Layout(width='90%', height='180px')
)

btn_process = widgets.Button(
    description='Run Mind Filter',
    button_style='success',
    tooltip='Click to execute Llama-3 processing',
    icon='bolt'
)

out_display = widgets.Output(
    layout=widgets.Layout(border='1px solid #ccc', padding='15px', margin='10px 0', width='90%')
)

def on_button_clicked(b):
    """
    Button trigger event handler. Handles UI switching and rendering.
    """
    if not txt_input.value.strip():
        with out_display:
            clear_output()
            print("Error: Input text area cannot be empty.")
        return

    btn_process.disabled = True
    btn_process.description = 'Processing via Llama-3...'

    with out_display:
        clear_output()
        print("AI is organizing your text structure. Please wait...")
        try:
            result = process_text_with_llm(txt_input.value)
            clear_output()
            # Corrected: Use IPython's Markdown class instead of widgets.Markdown
            display(Markdown(result))
        except Exception as e:
            clear_output()
            print(f"Runtime Error occurred: {str(e)}")

    btn_process.disabled = False
    btn_process.description = 'Run Mind Filter'

# Bind event handler to button
btn_process.on_click(on_button_clicked)

# Render UI layout on screen
# Corrected: Use IPython's Markdown class instead of widgets.Markdown
display(Markdown("## 🧠 Mind Filter Native Interface"))
display(txt_input)
display(btn_process)
display(out_display)
```
